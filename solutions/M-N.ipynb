```python
import pandas as pd

# 1. Загрузка данных
data = pd.read_csv('../synthetic_gmv_data_1.1.csv')

# 2. Агрегация данных по пользователям (сумма GMV)
user_aggregated = data.groupby('user_id')['gmv'].sum().reset_index()
print("Агрегированные данные по пользователям:")
print(user_aggregated.head())

# 3. Расчет выборочного среднего (несмещенная оценка матожидания)
sample_mean = user_aggregated['gmv'].mean()
print(f"\nВыборочное среднее (матожидание): {sample_mean:.3f}")

# 4. Расчет выборочной дисперсии (несмещенная оценка)
sample_variance = user_aggregated['gmv'].var(ddof=1)  # ddof=1 для несмещенной оценки
print(f"Выборочная дисперсия: {sample_variance:.3f}")

# Дополнительно: стандартное отклонение
sample_std = user_aggregated['gmv'].std(ddof=1)
print(f"Выборочное стандартное отклонение: {sample_std:.3f}")
```

```python
e = 0.01
a = 0.05
b = 0.2
k = 3
import scipy.stats as stats
import math
def size_mde(var, effect, k = 1, alpha = 0.05, beta = 0.2):
    dist = stats.norm(loc=0, scale=1)
    const = (dist.ppf(1- alpha/2) + dist.ppf(1- beta))**2
    return (const*(1+k)*var/(effect**2))

m = math.ceil(size_mde(sample_var, e * sample_mean, 3))
n = m / k
total_size = n + m
```
